# 02 - Primer directed mutagenesis with pydna



This workflow describes how to design and simulate a primer-directed mutagenesis library using PyDNA. The aim is to introduce predefined mutations into a protein-coding sequence, simulate the PCRs required for each variant, and assemble the resulting plasmids or genomic constructs in silico before ordering reagents.

This approach enables the efficient design of large-scale mutagenesis experiments while reducing experimental trial and error.

### Steps

1. Prepare a list of mutations

Create a list of desired amino acid or nucleotide mutations for your target protein (protein X).
These can include point mutations, insertions, or deletions — depending on your experimental goal (e.g., improving stability, activity, or substrate specificity).

2. Generate mutagenic primers

Design primers that introduce each mutation using a mutagenesis scheme such as site-directed PCR or overlap extension PCR.
- Ensure primer melting temperatures and GC content are suitable.

- For libraries, automate primer generation to cover all target variants.

3. Simulate PCR reactions (with pydna)

Use PyDNA to simulate the amplification steps for each mutagenesis reaction.

- Confirm that the expected amplicon sizes and mutation sites are correct.

- Verify that overlapping regions are compatible for downstream assembly.

4. Simulate genome integration or plasmid assembly
Combine PCR fragments and simulate assembly into a plasmid or genome locus.

Test both one-step and multi-step assembly options in pydna.

Validate that all constructs contain the intended mutations and are free of sequence conflicts.

5. Order all required materials
Once all designs are validated, order primers, gene fragments, and plasmids from your vendor (e.g., IDT).
This includes both mutation-specific primers and any standardized backbone components.
