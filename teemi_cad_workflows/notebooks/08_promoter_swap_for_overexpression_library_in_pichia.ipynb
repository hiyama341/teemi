# 08 - Promoter Swap for an Overexpression Library in Pichia


This workflow outlines how to design a promoter-swap overexpression library in Pichia pastoris or another methylotrophic yeast. The goal is to replace native promoters or install alternative strong promoters upstream of target genes in order to systematically tune expression across a set of strains. The process involves selecting promoter parts, identifying genomic insertion sites, designing HDR templates and guide RNAs, and simulating genome editing events with pydna.

A promoter-swap library is a strong way to probe pathway bottlenecks and expression sensitivity without changing coding sequences. In Pichia, this is especially attractive because promoter choice can strongly affect induction behavior, burden, methanol dependence, and productivity. The notebook should therefore help compare a panel of constitutive and inducible promoters across a defined set of genomic targets while keeping build logic and strain naming consistent.

### Steps

1. Select target genes for overexpression

- Choose the genes you want to place under stronger or differently regulated promoters.
- Define whether the library should target one pathway, one enzyme family, or a broader set of loci.
- Rank the targets if the total design space becomes too large to build all combinations at once.

2. Collect promoter sequences

- Assemble a panel of constitutive, inducible, or hybrid promoters suitable for Pichia.
- Store them in FASTA or GenBank format so they can be reused in combinatorial designs.
- Include metadata such as promoter type, expected strength, induction regime, and literature source when available.
- It is often useful to include both benchmark promoters and more exploratory candidates.

3. Obtain the genome context for each target

- Load a reference genome or locus sequence set for the Pichia strain of interest.
- Extract the upstream regions around each target gene to define promoter replacement boundaries.
- Review neighboring features so the promoter swap does not unintentionally disturb adjacent genes or regulatory elements.

4. Design promoter-swap repair templates

- Build HDR templates that replace the native promoter with each library promoter.
- Keep the coding sequence unchanged while preserving correct spacing to the start codon.
- Decide whether the swap is scarless or whether small synthetic junctions are acceptable.
- If multiple genes are targeted, standardize the repair template architecture as much as possible.

5. Identify CRISPR guides near the promoter region

- Search for guide RNAs that cut close to the promoter replacement site.
- Filter guides for efficiency and minimize obvious off-target liabilities.
- Favor guides that support precise replacement of the upstream region without damaging the coding sequence.
- If promoter lengths differ substantially, check that the chosen cut site still works cleanly across designs.

6. Simulate editing and assembly

- Use pydna to simulate the CRISPR cut and integration of each promoter-swap template.
- Confirm that the new promoter is installed in the correct orientation and context.
- Verify that the final design preserves the ORF, upstream architecture, and any required untranslated region features.
- If the project includes several promoters per target, produce a clear combinatorial map of the full strain set.

7. Design validation primers and screening strategy

- Generate PCR primers that distinguish native and promoter-swapped loci.
- Generate PCR primers that distinguish native and promoter-swapped loci.
- Consider how expression will actually be measured, for example transcript level, enzyme activity, growth phenotype, or reporter signal.
- Optionally add reporter tags or barcodes if the library will be screened in pooled format.
- If the library is large, define a naming convention that encodes both the target gene and the installed promoter.

8. Export the full overexpression library design

- Prepare ordering tables for guides, primers, donor templates, and any shared plasmid parts before moving to the wet lab.
- A useful final output is a strain design table that links each promoter swap to its intended phenotype and validation strategy.
