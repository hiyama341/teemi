# 06 - CRISPR Multiplex Editing Planner


This workflow outlines how to plan a multiplex CRISPR editing experiment where several genomic loci are edited in parallel. The goal is to identify guides, design donor DNA or repair templates, simulate edits, and prepare validation primers for a coordinated genome engineering campaign. The process combines sequence retrieval, guide filtering, donor design, and in silico validation using teemi, pydna, and Biopython.

A useful multiplex planner does more than list guides. It should connect each target locus to a specific edit objective, a compatible guide, a donor design strategy, and a validation plan. For example, one project might combine two pathway knockouts, one promoter replacement, and one reporter insertion in the same strain background. Planning those edits together helps reveal collisions in marker usage, repeated sequence elements, or donor designs that are difficult to distinguish during colony screening.

### Steps

1. Define the editing campaign

- List the genes or loci to delete, replace, activate, repress, or tag.
- Specify the edit type for each target so donor design and validation can be standardized.
- Record the biological rationale for each edit so the final strain set remains interpretable.
- Decide whether all edits are intended to occur in a single transformation or in staged rounds.

2. Load the reference genome and annotations

- Import the GenBank file for the host organism.
- Use locus tags and feature annotations to locate all target regions precisely.
- Verify that the genome version matches the strain actually used in the lab, especially if local polymorphisms are known.

3. Identify candidate guides for each locus

- Search for CRISPR target sites near every edit position.
- Score or filter guides based on PAM compatibility, strand, position, and off-target risk.
- Prefer guides that cut close enough to the desired edit to support efficient HDR.
- For multiplex designs, also consider whether guides can be expressed in a compatible array or cassette architecture.

4. Design donor templates or repair fragments

- Create HDR templates with homology arms tailored to each edit.
- For insertions or replacements, include the desired cassette, barcode, or marker between the arms.
- Ensure donor designs do not recreate the guide target site unless that is intentional.
- Standardize donor architecture when possible so synthesis, PCR setup, and documentation stay manageable.

5. Simulate multiplex editing outcomes

- Use pydna to simulate double-strand breaks and homologous recombination for each locus.
- Track expected edited sequences and check for conflicts between edits.
- Consider whether any pair of edits could interfere through repeated homology, shared markers, or viability effects.
- If the design includes alternative edit combinations, define those strain genotypes explicitly.

6. Design PCR and genotyping primers

- Generate diagnostic primers for each edited locus.
- Include wild-type versus edited amplicon expectations to simplify colony screening.
- Add a strategy for distinguishing partial edits from complete edits when multiple loci are modified at once.
- If sequencing validation is planned, note which amplicons should be submitted for Sanger or long-read confirmation.

7. Organize the experiment for execution

- Export guide tables, donor templates, primer sheets, and construct maps.
- Define which edits share reagents and which need unique reagents.
- If needed, prepare plate-ready layouts for pooled cloning or transformation workflows.
- A strong output here is a build sheet that links each strain design to its guide set, donor set, and validation primers.

8. Order all reagents

- Once all multiplex edits are validated in silico, order guides, repair templates, primers, and plasmid parts from your preferred supplier.
- Before ordering, perform one final review for guide uniqueness, donor naming consistency, and screening feasibility.
