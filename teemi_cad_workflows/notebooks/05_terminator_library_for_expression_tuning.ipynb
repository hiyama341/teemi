# 05 - Terminator Library for Expression Tuning


This workflow outlines how to design and construct a terminator library to tune gene expression in yeast or another expression host. The goal is to hold the promoter and coding sequence constant while varying transcription terminators to explore changes in mRNA stability, transcription termination efficiency, transcript processing, and protein output. The process involves selecting a measurable reporter system, retrieving a panel of terminators, defining a standardized expression cassette, and then designing and simulating the resulting library with pydna before any DNA is ordered.

A terminator-focused library is attractive because it isolates a regulatory variable that is often overlooked compared with promoter engineering. In many systems, changing only the terminator can alter expression strength, noise, burden, and transcript stability without changing the ORF or promoter. That makes the library easier to interpret and useful both for basic characterization and for production strain optimization.

### Steps

1. Select a reporter or production gene

- Choose a coding sequence to place under a fixed promoter.
- This can be a fluorescent reporter, an enzyme, or another measurable output gene.
- Prefer a readout that is cheap and fast to quantify so the entire terminator panel can be screened in parallel.
- Decide early whether the goal is expression benchmarking, secretion optimization, pathway balancing, or strain engineering.

2. Collect terminator sequences

- Gather a panel of native or synthetic terminators from the organism of interest.
- Include both strong and weak candidates if you want a broad dynamic range rather than only top-performing parts.
- Store them in FASTA or GenBank format for downstream processing.
- Keep track of sequence length, source gene, and any literature annotation that may help interpret later results.

3. Define the expression cassette architecture

- Standardize the promoter, coding sequence, linker or tag, and integration locus or plasmid backbone.
- Keep everything fixed except the terminator so the library isolates terminator effects.
- Decide whether the constructs will be built on plasmids or integrated into a common genomic landing pad.
- If genomic integration is used, select one locus for the whole library to minimize position effects.

4. Enumerate the construct library

- Use teemi to generate all promoter-CDS-terminator combinations of interest.
- In the simplest version, only the terminator changes. In a larger design, you may also include one or two promoters or tags as controlled factors.
- Add systematic names so each design can be tracked through assembly, strain construction, and screening.
- Create a sample sheet that maps construct names to terminator identities and expected phenotypes.

5. Design primers for assembly

- Generate primers for amplifying or synthesizing each terminator and any shared backbone parts.
- Reuse standardized primers when possible to reduce ordering cost.
- Check that overlaps or homology arms remain compatible across all variants.
- If some terminators are very similar, make sure naming and primer tracking are unambiguous before ordering.

6. Simulate cloning or HDR integration

- Use pydna to assemble each expression cassette in silico.
- Confirm that the terminator sits directly downstream of the coding sequence in the correct orientation.
- If integrating genomically, simulate CRISPR-mediated insertion at the chosen locus.
- Watch for sequence-context problems such as repeated parts, duplicated overlaps, or problematic junctions.

7. Prepare screening-ready outputs

- Export primer tables, construct summaries, and expected assembly products.
- Define the screening assay in advance, for example fluorescence, growth-coupled output, enzyme activity, or qPCR.
- Optionally add genotyping primers for validating correct integration of each terminator variant.
- If the library is large, prepare plate maps and well-level naming conventions so build and test data can be merged later.

8. Order oligos and DNA parts

- Once the library is validated in silico, order primers, terminator fragments, and any backbone components from your preferred vendor.
- At this stage the notebook should ideally produce a final list of parts, a design summary table, and a clear plan for downstream characterization.
