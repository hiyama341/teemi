## Paper 2: [A Mad7 System for Genetic Engineering of Filamentous Fungi](https://pmc.ncbi.nlm.nih.gov/articles/PMC9865164/)


**Objective:**
Recreate [pDIV298](https://www.addgene.org/204957/sequences/)

What they did: 
- Bought Ptef1::ErCas12a::Ttef1 from GeneArt and constructed pDIV298 by inserting a PCR fragment containing the Ptef1::ErCas12a::Ttef1 into USER cassette of [pAC572](https://www.addgene.org/127973/)


**Overview of the main plasmid in the paper**;

| ID     | Plasmid  | Description |
|--------|---------|-------------|
| 204957 | pDIV298 | AMA1 plasmid with *Aspergillus*-optimized Mad7 and *pyrG* selection marker |



### Matrials and methods text

The ErCas12a gene from Eubacterium rectale [20] was codon-optimized for translation in A. niger and synthetized by GeneArt™ (Thermo Fisher Scientific). USER compatible backbone encoding the ErCas12a gene was created by fusing it with A. nidulans tef1 promoter and terminator and then by incorporating into a digested USER vector backbone. 


The primer fusing the tef1 promoter with the digested backbone included a new USER restriction site (PacI/Nt.BbvCI USER cassette). A total of six Mad7-CRISPR vectors with different selective markers: pDIV298 (pyrG, selection/counterselection by uracil + uridine prototrophy/5-fluoroorotic acid), pDIV299 (argB, selection by arginine prototrophy), pDIV300 (hygB, selection by hygromycin resistance), pDIV301 (ble, selection by bleomycin resistance), pDIV302 (nat, selection by nourseothricin resistance) and pDIV303 (amdS, selection/counterselection by acetamide prototrophy/fluoroacetamide) **were constructed by inserting a PCR fragment containing the Ptef1::ErCas12a::Ttef1 into USER cassette of pAC572, pAC573, pAC574, pAC575, pDIV296 and pDIV297, respectively (see Table S2)**. 

The PCR fragment containing the ErCas12a gene was generated by primers PR_DIV1503 and PR_DIV1504 using the synthetic gene from GeneArt™ as a template. Newly created Mad7-CRISPR USER compatible backbones were used in subsequent experiments to create vectors encoding the gRNA expression cassette. The gRNA cassette was constructed fusing two PCR products by USER fusion, see Figure S1.

```python
from pydna.primer import Primer
from pydna.dseqrecord import Dseqrecord
```

```python
PR_DIV1503 = Primer('ACCTTCACCAUATGAACAACGGCACCAACAAC') # original seq ACCTTCACCAUATGAACAACGGCACCAACAAC
PR_DIV1504 = Primer('ATGTCCGCUTCAGACCTTGCGCTTCTTCTT')
```

```python
from Bio import SeqIO
record = Dseqrecord(SeqIO.read("data/addgene-plasmid-204957-sequence-415603 (1).gbk", "genbank"), circular = True)
```

```python
from pydna.amplify import pcr
mad7 = pcr(PR_DIV1503,PR_DIV1504, record)
print(mad7)
```

```python
from teemi.design.cloning import USER_enzyme
mad7_USER = USER_enzyme(mad7)
print(mad7_USER.figure())
```

```python
mad7_USER.seq.crick
```

```python
from Bio.Restriction import PacI
backbone_plasmid = Dseqrecord(SeqIO.read("data/127973-attachment_lFglSbDyj20P.gbk", "genbank"),circular=True)
backbone_plasmid_PacI  = backbone_plasmid.linearize(PacI)
print(backbone_plasmid_PacI)
```

```python
from teemi.design.cloning import Nt_Bbc_CI
backbone_plasmid_PacI_nick = Dseqrecord(Nt_Bbc_CI(backbone_plasmid_PacI))
print(backbone_plasmid_PacI_nick)
```

```python
# 3.1: Combine the nicked vector with the USER processed gRNAs and loop the resulting sequence
pDIV298 =  (backbone_plasmid_PacI_nick + mad7_USER).looped()
pDIV298.seq
```
